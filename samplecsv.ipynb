```
df=spark.read.csv('/FileStore/tables/sales_data_sample.csv',header=True)
df.show()
```

```
import pandas as pd
pddf=df.toPandas()
pddf['ORDERDATE'] = pd.to_datetime(pddf.ORDERDATE).dt.strftime('%m/%d/%y')
#datetime.strftime(ti, '%y/%m/%d')
df_sp = spark.createDataFrame(pddf)
```

```
df_sp.show()
```

```
add_df_sp=df_sp.withColumn('ORDERYEAR',df_sp.ORDERDATE.substr( 7,8))
add_df_sp.show()
```
