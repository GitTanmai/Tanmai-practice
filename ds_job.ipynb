```
from collections import Counter 
from collections import OrderedDict
from nltk.corpus import stopwords


df_panda=deslist_df.select("*").toPandas()
desc_listing=list(df_panda['Description'])
```

```
def desc_counter(cnt):
  stop_words = stopwords.words('english')
  allst=""
  stop_words = stopwords.words('english')
  for d in cnt:
    allst += d
  allst=allst.lower()
  alllist = re.findall('[\w]+',allst)
  #list(set(x) - set(stop_words))
  cnt=Counter(x for x in alllist if x not in stop_words)
  #cnt=Counter(alllist)
  return cnt
```

```
alld= desc_counter(desc_listing)
most_comlist=alld.most_common()
dd=dict(most_comlist)
dd
```
